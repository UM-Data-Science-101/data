# Exploring the NBA_API Library

The nba_api is a powerful Python library that provides programmatic access to NBA data sourced directly from NBA.com. Developed and maintained by the community, it allows users to retrieve a wide range of statistics, game logs, player profiles, team data, shot charts, and advanced metrics through a clean and flexible interface. The library is especially popular among data scientists and basketball analysts for building dashboards, conducting performance analysis, and training predictive models.

```python
from nba_api.stats.endpoints import leaguedashplayerstats
import pandas as pd
```

## Player stats

```python
# Get player stats for the 2024–25 season
stats = leaguedashplayerstats.LeagueDashPlayerStats(season='2024-25')
df = stats.get_data_frames()[0]

# Preview the data
df.head()
```

## Logs for specific players

```python
from nba_api.stats.static import players

all_players = players.get_players()

# Search for a player
lebron = [p for p in all_players if p['full_name'] == 'LeBron James'][0]

# LeBron James Player ID
lebron_id = lebron['id']

print(lebron_id)
```

```python
from nba_api.stats.endpoints import playergamelog

# Get game logs for the 2024-25 season for Lebron
log = playergamelog.PlayerGameLog(player_id=lebron_id, season='2024-25')
df_log = log.get_data_frames()[0]
df_log[['GAME_DATE', 'PTS', 'REB', 'AST']]
```

## Teams

```python
from nba_api.stats.static import teams

# List all NBA teams
nba_teams = teams.get_teams()
teams_df = pd.DataFrame(nba_teams)
teams_df.head()
```
